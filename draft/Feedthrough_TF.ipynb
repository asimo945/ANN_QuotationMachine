```python
import numpy as np
import pandas as pd
import tensorflow as tf
from TF_data_prepro import*
```

```python
## load data ##
path=r'df_clean.pkl'
df= pd.read_pickle(path)
```

```python
## set target ##
pd.set_option('display.max_columns', None)
df['target']=df['price per unit']
df=df.drop(columns='price per unit')
df.reset_index(inplace=True,drop=True)
df
```

```python
df,num,cat=feature_col_clean_split(df)
num.head()
```

```python
cat.head()
```

```python
train, test, val=split_data(df, 0.2)
```

```python
[(train_features, label_batch)] = train_ds.take(1)
print('Every feature:', list(train_features.keys()))
print('A batch of order quantity:', train_features['order quantity'])
print('A batch of targets:', label_batch )
```

```python
feature_columns = []
```

```python
## numerical cols ##
for header in num.columns[:-1]:
    scal_input_fn = get_scal(header)
    feature_columns.append(feature_column.numeric_column(header, normalizer_fn=scal_input_fn))
```

```python
feature_columns
```

```python
for feature_name in cat.columns:
    vocabulary = cat[feature_name].unique()
    cat_c = tf.feature_column.categorical_column_with_vocabulary_list(feature_name, vocabulary)
    embeding = feature_column.embedding_column(cat_c, dimension=50)
    feature_columns.append(embeding)
```

```python
feature_columns
```

```python
feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
```

```python
batch_size = 32
train_ds = df_to_dataset(train, batch_size=batch_size)
val_ds = df_to_dataset(val, shuffle=False, batch_size=batch_size)
test_ds = df_to_dataset(test, shuffle=False, batch_size=batch_size)
```

```python
model = tf.keras.Sequential([
  feature_layer,
  layers.Dense(128, activation='relu'),
  layers.Dense(128, activation='relu'),
  layers.Dropout(.1),
  layers.Dense(1)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.fit(train_ds,
          validation_data=val_ds,
          epochs=10)
```

```python
loss, accuracy = model.evaluate(test_ds)
print("Accuracy", accuracy)
```
